```python
Grouped = dfAssociation.groupby('itemid')["orderid"].apply(set)
```

```python
"""
Now Lets Find the Correspond Order id Y have item id X
"""

def conf(dat):
    dat = [item.split("&")for item in dat.split(">")]
    print(dat,"\n\n") # [[pembagi] , [numerator]]

    Pembagi = Grouped.loc[int(dat[0][0])]
    
    if len(dat[0]) > 1: # Jadi kalau ada And i.e &
        pembagi1 = int(dat[0][0])  
        pembagi2 = int(dat[0][1])

        Pembagi = Grouped.loc[pembagi1] & Grouped.loc[pembagi2] # lakuin Intersection  < Retrun order id yang ber irisan >
 
    Numerator = Pembagi
    for i in dat[1]:
        Numerator = Numerator & Grouped.loc[int(i)]

    return math.floor(len(Numerator) / len(Pembagi)* 1000) 


conf("1667608337>1693650517&2015582146")

```

```python
dfRules["confidence"] = dfRules["rule"].apply(conf)
dfRules.head()
```
